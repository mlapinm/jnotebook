## Алгоритмы поиска

### Последовательный (линейный) поиск

````python
def linear_search(lst, key):
    for idx, item in enumerate(lst):
        if key == item:
            return idx
    return -1
````

```
Элемент	  +    -    +/-
-----------------------------
Есть         1    n    n/2
Нет          n    n    n
```

````python
def ordered_linear_search(lst, key):
    for idx, item in enumerate(lst):
        if item == key:
            return idx
        elif item > key:
            return -1
        return -1
````

```
Элемент	  +    -    +/-
-----------------------------
Есть         1    n    n/2
Нет          n    n    n/2
```

### Бинарный поиск

````python
def binary_search(lst, key):
    first = 0
    last = len(lst)-1
    
    while first<=last:
        mid = (first + last) // 2
        if lst[mid] == key:
            return mid
        else:
            if key < lst[mid]:
                last = mid-1
            else:
                first = mid+1
    return -1
````

```
Элемент	  +          -       +/-
---------------------------------------
Есть         1        log n    log n
Нет          log n    log n    log n
```

# Выводы
#### Поиск является одной из важнейших процедур обработки структурированной информации 
#### Асимптотическая сложность алгоритма линейного поиска — O(n)
#### Асимптотическая сложность алгоритма бинарного поиска для упорядоченных списков — O(log n)
