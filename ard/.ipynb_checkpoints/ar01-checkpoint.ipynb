#### SOS

````python
const uint8_t PIN_LED = 13;
const uint8_t PIN_BTN = 2;
const uint32_t DOT = 100;
const uint32_t DASH = 500;
const uint32_t PAUSE = 500;



void setup() {
  pinMode(PIN_LED, OUTPUT);
  digitalWrite(PIN_LED,LOW);
  pinMode(PIN_BTN, INPUT_PULLUP);
  Serial.begin(115200);
}

void blink(uint32_t pulse, uint32_t pause){
  digitalWrite(PIN_LED, HIGH);
  delay(pulse);
  digitalWrite(PIN_LED, LOW);
  delay(pause);
}

void sos(){
  uint8_t i;

    for(i=0;i<3;i++){
      blink(DOT,PAUSE); 
      }
    delay(PAUSE);
    for(i=0;i<3;i++){
      blink(DASH,PAUSE); 
      }
    delay(PAUSE);
    for(i=0;i<3;i++){
      blink(DOT,PAUSE); 
      }
}

void loop() {
  if(!digitalRead(PIN_BTN)){
    sos();
    }
}
````

#### Кнопка, дребезг, 2 режима

````python
const uint8_t LED_PIN = LED_BUILTIN;
const uint8_t BTN_PIN = 2;

void setup() {
  pinMode(LED_PIN, OUTPUT);
  pinMode(BTN_PIN, INPUT_PULLUP);
  
  Serial.begin(115200);
  Serial.println();

  Serial.println(LED_PIN);

}

void loop() {
  static const uint32_t LED_TIME = 5000;

  static bool lastBtn = false;
  static uint32_t ledDuration = 0;

  bool btn = debounceRead(BTN_PIN, LOW);
  if(btn != lastBtn){
  Serial.print(lastBtn);
  Serial.println(btn);
    lastBtn=btn;
    if(btn){
      if(ledDuration){
        digitalWrite(LED_PIN, LOW);
        ledDuration = 0;
      }else{
        digitalWrite(LED_PIN, HIGH);
        ledDuration = millis()+ LED_TIME;
      }
    }  
  }
  if(ledDuration){
    if((int32_t)millis()>(int32_t)ledDuration){
      digitalWrite(LED_PIN, LOW);
      ledDuration = 0;  
    }
  }
}

bool debounceRead(uint8_t pin, bool level){
  static const uint32_t DEBOUNCE_TIME = 20;

  static bool lastBtn = false;
  static uint32_t lastTime;

  bool btn = digitalRead(pin) == level;
  if(btn != lastBtn){
    lastBtn = btn;
    lastTime = millis();  
  }else{
    if(millis()-lastTime >= DEBOUNCE_TIME){
      return btn;  
    }
  }
  return false;
}
````

````python
#include <string.h>
#include <avr/pgmspace.h>

enum sex_t : uint8_t { MALE, FEMALE};

struct person_t{
  char name[30];
  uint8_t age:7;
  sex_t sex:1;
};

void setPerson(person_t &person, const char *name,uint8_t age,sex_t sex){
  strncpy(person.name,name,sizeof(person.name)-1);
  person.name[sizeof(person.name)-1] = '\0';
  person.age = age;
  person.sex = sex;  
}

void setPerson(person_t &person, const __FlashStringHelper *name, uint8_t age,sex_t sex){
  strncpy_P(person.name,(PGM_P)name,sizeof(person.name)-1);
  person.name[sizeof(person.name)-1] = '\0';
  person.age = age;
  person.sex = sex;  
}


void setup() {
  Serial.begin(115200);
  Serial.println();
  person_t alex;
  person_t daria;
  
  setPerson(alex,F("Alex"),44,MALE);
  setPerson(daria,F("Daria"),6,FEMALE);

  printPerson(alex);
  printPerson(daria);
}

void loop() {}

void printPerson(person_t &person){
  Serial.print(F("Hi, "));
  Serial.println(person.name);
  if(person.sex == FEMALE){
    Serial.print(F("fe"));
    }
  Serial.print(F("male "));
  Serial.print(F("Your age is "));
  Serial.println(person.age);
}
````
