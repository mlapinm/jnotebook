https://htmlweb.ru/php/php_regexp.php  
Упорядочивает результаты так, что элемент \\$matches[0] содержит массив полных вхождений шаблона, элемент \\$matches[1] содержит массив вхождений первой подмаски и т.д.  
"|<[^>]+>(.*)</[^>]+>|U"
```
"|<[^>]+>(.*)</[^>]+>|U"
| - ограничитель re (м.б. ~)
Начинается с < , дальше [^>]+ - любой символ кроме >, повторяющийся 1 и более раз.
(.*) - все от нуля и больше.
U - не жадный, ищет первый, а не последний
```
