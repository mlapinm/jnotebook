Кластеризация:
High availability;
Load balancing
```

VRRP:  
Virtual Router Redundancy Protocol — протокол, предназначенный для увеличения доступности маршрутизаторов выполняющих роль шлюза по умолчанию. Выполняет объединение группы маршрутизаторов в один виртуальный маршрутизатор и назначает им общий 1Р-адрес.

```
Параметры VRRP:
Итерфейс;
Идентификатор VR;
Приоритет;
Preemption.
```

```
Пример-Переключение:
В случае отказа Router со статусом Master, происходит захват роли Backup маршрутизатором с наивысшим приоритетом.
```

```
VRID= 100
R1 int eth2 ip = 192.168.10.252

vIP = 192.168.10.254

Состояния:

Разделения устройств в группы, 
Предотвращение широковещательных штормов, 
Увеличения безопасности.

Порты:
По умолчанию все порты коммутатора находятся во VLAN 1; 
VLAN разделяют коммутатор на широковещательные домены.

Типы портов:
Нетегированные (Access); 
Тегированные (Trunk).

Маршрутизация:
Физический интерфейс на каждую подсеть;
Один физический интерфейс с вложенными sub-интерфейсами для каждой подсети
