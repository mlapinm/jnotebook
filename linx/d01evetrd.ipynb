```
Способы использования компьютеров в UNL:
1. VPCS
2. VMware Workstation
3. Свой компьютер или реальные компьютеры из сети
4. QEMU
5. Docker
Плюсы использования QEMU
1. Вся схема внутри UNetLab
2. Одну виртуальную машину можно использовать в схеме множество раз
Минусы использования QEMU
Сложная и запутанная настройка
Возможны два варианта использования QEMU
1. Для запуска Live дистрибутивов
2. Для запуска установленных операционных систем
```

/opt/unetlab/html/includes/init.php

```
Алгоритм создания виртуальной машины qemu
1. В папке /opt/unetlab/addons/qemu создаем папку согласно файлу /opt/unetlab/html/includes/init.php
(пример: win-xp, linux-kali)
2. Закидываем в эту папку образ и переименовываем его в cdrom.iso (для live дистрибутивов на этом
настройка закончена)
3. Создаем жесткий диск командой /opt/qemu-2.0.2/bin/qemu-img create -f qcow2 hda.qcow2 3000M
4. Задаем права командой /opt/unetlab/wrappers/unl_wrapper -a fixpermissions
5. Стартуем и сразу стопим ноду
6. Копируем жесткий диск (hda.qcow2) в соответствующую папку /opt/unetlab/tmp
7. Производим установку
8. По завершению копируем жесткий диск из /opt/unetlab/tmp обратно в
/opt/unetlab/addons/qemu/ваша_папка
9. Зачищаем /opt/unetlab/tmp
10. Задаем права /opt/unetlab/wrappers/unl_wrapper -a fixpermissions
```

#### 9.UNetLab. Компьютер в сетевой топологии - Docker


```

```

1) Полная (аппаратная) виртуализация 

2) Виртуализация уровня ОС 

3) Виртуализация приложений

[350+ полезных ресурсов, книг и инструментов для работы с Docker](https://habr.com/ru/company/1cloud/blog/275015/)  
[BusyBox](https://ru.wikipedia.org/wiki/BusyBox)  
[]()  


**Настройка интеграции Docker с UNetLab:**  
`/etc/default/docker строчку DOCKER_OPTS="-H=tcp://0.0.0.0:4243"` - дописать в файл  
**Управление образами в Docker:**  
`docker -H=tcp://127.0.0.1:4243 search ubuntu` - Поиск обрзов  
`docker -H=tcp://127.0.0.1:4243 pull ubuntu:latest` - Скачивание образов  
`docker -H=tcp://127.0.0.1:4243 images` - Просмотр образов  
**Настройка сети осуществляется через startup-configs:**  
`ip addr add 1.1.1.1/24 dev eth0` -   
`ip route add default via 1.1.1.254` -   
В настройке ноды в поле sturtup-config должно быть значение Exported   
`` -   


#### 10.UNetLab. Cisco vIOS


Cisco VIRL - эмуляция IOSv, IOS XRv, NX-OSv, CSR1000 и ASAv  
(all-in-one-VM-1.2.0-173.ova)  
(all-in-one-VM-1.3.0-181.ova)  
**vIOS L2 (коммутатор)**   
- Layer-2 forwarding
- Switchport
- 802.1q trunk, 802.1q vlans
- Spanning Tree
- Port-Channel (Pagp and Lacp)
- 802.1x passthrough
- Port-ACLs
- Dynamic Arp Inspection
- DHCP Snooping
- IP device tracking
- Switched Virtual Interfaces
- Layer-3 forwarding over SVIs
- Routing protocols
Чего НЕ умеет данный коммутатор:
- Зеркалирование трафика (SPAN)
- Private Vlan
- L3 Port-Chennel
- QOS  
**vIOS (маршрутизатор)**  
- Стандартный набор функций

##### Найти данные образы можно в интернете используя такие запросы как:
“iosv-l2 download”, “iosv-l3 download”
##### Алгоритм добавления образов:
1. В папке /opt/unetlab/addons/qemu создаем папки согласно файлу /opt/unetlab/html/includes/init.php
(пример: vios-l3-15.4, viosl2-15.0)
2. Закидываем в соответствующие папки образы и переконвертируем их в формат qemu. Пример:
Для vIOS L2 - /opt/qemu/bin/qemu-img convert -f vmdk -O qcow2 vIOS-L2.vmdk virtioa.qcow2
Для vIOS L3 - /opt/qemu/bin/qemu-img convert -f vmdk -O qcow2 vIOS-L3.vmdk virtioa.qcow2
3. Задаем права командой /opt/unetlab/wrappers/unl_wrapper -a fixpermissions
4. При добавлении устройств выставляйте 384 Мб оперативной памяти.
5. Для коммутатора максимальное количество портов - 16.

#### 11.UNetLab. Cisco ASA

##### Cisco ASA - самый известный межсетевой экран (16-й урок Курса молодого бойца)
##### В UNetLab доступно два вида Cisco ASA:
**1) Cisco ASA 8.0.2, 8.4.2, 9.1.15 (Support Multi Context)**  
Для поиска образов использовать следующие запросы в google: “ASA842-disk1.vmdk” и “ASA-9.15-disk1.vmdk”  
Системные требования: 512-768 Mb RAM и 100% загрузка ядра.  
При добавлении в папке /opt/unetlab/addons/qemu создается папка asa-... (asa-8.4 или asa-9.1)  
**2) Cisco ASAv (ver 9.3.2.200; 9.4.1; 9.5.1)**  
Для поиска образов использовать следующие запросы в google: “asav952-204.qcow2” или “asav941-200.qcow2”  
Системные требования: 1024-2048 Mb RAM и незначительная загрузка ядра.  
При добавлении в папке /opt/unetlab/addons/qemu создается папка asav-... (asav-9.4 или asav-9.5)  
*Алгоритм добавления образов:*  
1. В папке `/opt/unetlab/addons/qemu` создаем папки согласно файлу   `/opt/unetlab/html/includes/init.php` (пример: asa-8.4, asa-9.1,  
asav-9.4 или asav-9.5)
2. Закидываем в соответствующие папки образы и переконвертируем их в формат   qemu. Пример:  
Для ASA - `/opt/qemu/bin/qemu-img convert -f vmdk -O qcow2 ASA842-disk1.vmdk hda.qcow2`  
Для ASAv - достаточно переименовать образ в `virtioa.qcow2`  
3. Задаем права командой `/opt/unetlab/wrappers/unl_wrapper -a fixpermissions`  

##### 1.Курс UNetLab. Введение
##### 2.Курс UNetLab. Установка
##### 3.Курс UNetLab. Dynamips
##### 4.UNetLab. Telnet, VNC, Wireshark
##### 5.UNetLab. Создание сетевой топологии
##### 6.UNetLab. Компьютер в сетевой топологии - VPCS
##### 7.UNetLab. Компьютер в сетевой топологии - VMware
##### 8.UNetLab. Компьютер в сетевой топологии - QEMU
##### 9.UNetLab. Компьютер в сетевой топологии - Docker
##### 10.UNetLab. Cisco vIOS
##### 11.UNetLab. Cisco ASA
