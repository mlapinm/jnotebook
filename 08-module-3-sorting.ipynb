## Алгоритмы сортировки

### Сортировка пузырьком

```python
def bubble_sort(lst):
    
    for cnt in range(len(lst)-1, 0, -1):
        
        for i in range(cnt):
            
            if lst[i] > lst[i+1]:
                
                lst[i], lst[i+1] = lst[i+1], lst[i]
```

### Сортировка вставками 

```python
def insertion_sort(lst):
    
    for i in range(1, len(lst)):
        
        current = lst[i]
        pos = i

        while pos>0 and lst[pos-1]>current:
            
            lst[pos] = lst[pos-1]
            pos = pos-1

        lst[pos] = current
```
