```python
def mReadFile(fileName):
    s = ''
    with open(fileName, encoding="utf8") as file:
        s = file.read()
    return s

def mWriteFile(fileName, text):
    with open(fileName, "w", encoding="utf8") as file:
        file.write(text)
    

def mStrToList(s):
    l = s.split('\n')
    l2 = []
    for v in l:
        for v2 in v.split(' '):
            l2.append(v2.strip())
    #     print(v)
        pass
    l2 = list(filter(lambda v: len(v.strip())>0, l2))
#     print('l2', l2)
    return l2
```
