````python
import re

def mReadFile(fileName):
    s = ''
    with open(fileName, encoding="utf8") as file:
        s = file.read()
    return s

def mWriteFile(fileName, text):
    with open(fileName, "w", encoding="utf8") as file:
        file.write(text)
    

def mStrToList(s):
    l = s.split('\n')
    l2 = []
    for v in l:
        l2.append(v.strip())
        pass
    l2 = list(filter(lambda v: len(v.strip())>0, l2))
#     print('l2', l2)
    return l2
print('ok')

def replaceSpace(s):
    s2 = ''
    sr = '+():#.'
    s = s.strip()
    s = s.replace(' ','-')
    for e in s:
        if sr.find(e) != -1:
            pass
        else:
            s2 += e
    return s2

````

````python
import re

def codeListing(s1, l2):

    l1 = mStrToList(s1)
    
    l01 = []
    for e in l1:
        match = re.search('\\\([^\\\]*)$', e)
        if match and len(match.groups())>0:
            e2 = match.groups()[0]
            l01.append(e2)
        else:
            l01.append(e)
        pass
    
    l12 = []
    for e in l01:
        e2 = ['', '']
        e2[0] = e
        e2[1] = replaceSpace(e)
        l12.append(e2)
        pass
        
#     l12 = l1
    
    l21 = []
    k = 1
    for e in l12:
        e2 = f'{k} [{e[0]}](#{e[1]})   '
        l21.append(e2)
        k += 1
    
    l22 = []
    for e in l12:
        e2 = f'## {e[0]}'
        l22.append(e2)
        l22.append('```')
        l22.append('')
        l22.append('```')
        
    l20 = l21 + l22
    for e in l20:
        l2.append(e)
    return 

l2 = []
codeListing(s1, l2)
for e in l2:
    print(e)
    
````
