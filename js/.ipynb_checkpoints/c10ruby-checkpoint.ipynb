#### 2 If
```
if true
  puts "true"
end
```
#### 3 Else
```
if false
  puts "true"
else
  puts "false"
end
```
#### 4 Elsif
```
x = 2
if x > 2
  puts "x > 2"
elsif x > 1
  puts "x > 1"
else
  puts "x <= 1"
end
```
#### 5 Unless
```
hungry = false

unless hungry
  puts "I'm writing Ruby programs!"
else
  puts "Time to eat!"
end
```
#### 6 Equal or Not?
```
is_true = 2 != 3
is_false = 2 == 3
```
#### 7 Less Than or Greater Than
```
test_1 = 17 > 16

test_2 = 21 < 30

test_3 = 9 >= 9

test_4 = -11 <= 4
```
#### 8 Practice Makes Perfect
```
# test_1 = 77 != 77
test_1 = false

# test_2 = -4 <= -4
test_2 = true

# test_3 = -44 < -33
test_3 = true

# test_4 = 100 == 1000
test_4 = false
```
