#### 13. Sum All Odd Fibonacci Numbers
```
function fibs(value){
  if(value <= 1){
    return 1;
  }else{
    return fibs(value-1)+fibs(value-2);
  }  
}

function fibsArrLess(value){
var arr = [];
var f = 0;
var index = 0;
while(f <= value){
  f = fibs(index);
  if(f <= value){
    arr.push(f);
  }
  index++;
}
 return arr;
}

function sumFibs(num) {
  var arr = fibsArrLess(num);
  var arrOddSum = arr
  .filter(v => v % 2 != 0)
  .reduce((s, e) => s += e, 0);
  return arrOddSum;
}

console.log(sumFibs(4));
```
