````python
#coding: utf8
import time
import serial

port = 'COM3'
ser = serial.Serial(port, 9600, timeout=0.2)

#-- отправить строку и принять ответ
def fn_outin( st_out ):
    global ser 
    ser.write( st_out.encode() )
    st_answ = ser.readline().decode('cp1251');
    return  st_answ

#-- вывести строку в текстовое поле
def fn_print(st):
    tx_sect.insert(END,st[:-1] + '\n')

#-- нарисовать форму
from tkinter import *
# from tkFont import Font
from tkinter import font
root = Tk()
dFont = font.Font(family="helvetica", size=14)

#-- панель с органами управления
pn_left2 = Frame(root, width = 300, bg = 'darkblue')
pn_left2.pack(side = 'left', fill = 'y')

row_num = 1
cl_contr = 'darkblue'

#-- поле ввода отправляемой строки 
Label(pn_left2, text = 'строка для отправки:', bg = cl_contr, fg = 'white', font=dFont ).grid(row = row_num, column = 0, sticky=W, padx = 10)
row_num += 1

ed_str = Entry(pn_left2, width = 28, font=dFont)
ed_str.grid(row = row_num, column = 0, columnspan = 2, sticky=W, padx=10, pady = 10)
row_num += 1

#-- кнопка очистки строки ввода
def fn_cmdclear(event):
    ed_str.delete(0, END)

bt_clear = Button(pn_left2, text = 'очистить', width = 12, font=dFont)
bt_clear.bind("<Button-1>", fn_cmdclear)
bt_clear.grid(row = row_num, column = 0, sticky=W, pady = 2, padx = 10 )

#-- кнопка отправки введенной строки
def fn_send(event):
    st_out = ed_str.get().strip() +"\n"
    fn_print( st_out )
    st_in= fn_outin( st_out )
    fn_print( st_in)

bt_str = Button(pn_left2, text = 'отправить', width = 12, font=dFont)
bt_str.bind("<Button-1>", fn_send )
bt_str.grid(row = row_num, column = 1, sticky=W, pady = 2, padx = 10 )

ed_str.bind('<Return>', fn_send)

#-- правая панель для текста сектора
pn_right = Frame(root, width = 300)
pn_right.pack(side = 'right', fill = 'both', expand = 1)

#-- поле для вывода текста в правой панели
tx_sect = Text(pn_right, height = 16, width = 32, font=dFont)
tx_sect.grid(row = 0, column = 0)

sb_hor = Scrollbar(pn_right, command = tx_sect.xview, orient = HORIZONTAL)
sb_hor.grid(row = 1, column = 0, sticky = W+E+N+S)

sb_ver = Scrollbar(pn_right, command = tx_sect.yview, orient = VERTICAL)
sb_ver.grid(row = 0, column = 1, sticky = W+E+N+S)

tx_sect.config(yscrollcommand = sb_ver.set, xscrollcommand = sb_hor.set, wrap = NONE)

#-- запустить окно программы
root.mainloop()
````

#### py051 Простой таймер учета времени

````python
# -*- coding: utf-8 -*-
from tkinter import *
# from tkFont import Font
from tkinter import font
import winsound
from datetime import datetime,timedelta

root = Tk()
dFont = font.Font(family="helvetica", size=12)

#== настройки 
w_izmer = 400       #-- ширина панели
flag_work = 0       #-- таймер стартовал
flag_calc = 0       #-- таймер считает
flag_alarm = 0      #-- звонок включен

cl_izmer = '#008800'    #-- цвет формы
dWd = 8             #-- ширина поля вывода

#== панель с измерениями
root.title("Timer by AB")
pn_izmer = Frame(root, width = w_izmer, bg = cl_izmer)
pn_izmer.pack(side = 'left', fill = 'both')

#== первая строка:  время, кнопка "начать"
row_num = 0     #-- номер строки
lb_curtime = Label(pn_izmer, text = 'время:', bg = cl_izmer, fg = 'white', font = dFont )
lb_curtime.grid(row = row_num, column = 0, sticky=W, padx = 5)

ed_curtime = Entry(pn_izmer, width = dWd, font = dFont )
ed_curtime.grid(row = row_num, column = 1, padx= 5, pady= 5, sticky="nesw")

bt_start = Button(pn_izmer, text = u'начать', font = dFont )
bt_start.grid(row = row_num, column = 2, padx= 5, pady= 5, ipadx= 5, sticky="we")

lb_starttime = Label(pn_izmer, text = 'время старта:', bg = cl_izmer, fg = 'white', font = dFont )
lb_starttime.grid(row = row_num, column = 3, sticky=W, padx = 5)

ed_starttime = Entry(pn_izmer, width = dWd, font = dFont)
ed_starttime.grid(row = row_num, column = 4, padx= 5, pady= 5, sticky="nesw")

#== вторая строка: работа с интервалом
row_num += 1
lb_beginterval = Label(pn_izmer, text = 'начало:', bg = cl_izmer, fg = 'white', font = dFont )
lb_beginterval.grid(row = row_num, column = 0, sticky=W, padx = 5)

ed_beginterval = Entry(pn_izmer, width = dWd, font = dFont)
ed_beginterval.grid(row = row_num, column = 1, padx= 5, pady= 5, sticky="nesw")

bt_interval = Button(pn_izmer, text = u'остановить', font = dFont)
bt_interval.grid(row = row_num, column = 2, padx= 5, pady= 5, ipadx= 5, sticky="we")

lb_interval = Label(pn_izmer, text = 'интервал(мин):', bg = cl_izmer, fg = 'white', font = dFont )
lb_interval.grid(row = row_num, column = 3, sticky=W, padx = 5)

ed_interval = Entry(pn_izmer, width = dWd, font = dFont)
ed_interval.grid(row = row_num, column = 4, padx= 5, pady= 5, sticky="nesw")

#== третья строка: обработка звонка
row_num += 1
lb_alarm = Label(pn_izmer, text = 'звонок:', bg = cl_izmer, fg = 'white', font = dFont )
lb_alarm.grid(row = row_num, column = 0, sticky=W, padx = 5)

ed_alarm = Entry(pn_izmer, width = dWd, font = dFont)
ed_alarm.grid(row = row_num, column = 1, padx= 5, pady= 5, sticky="nesw")

bt_alarm = Button(pn_izmer, text = u'завести', font = dFont)
bt_alarm.grid(row = row_num, column = 2, padx= 5, pady= 5, ipadx= 5, sticky="we")

lb_alarminterval = Label(pn_izmer, text = 'интервал(мин):', bg = cl_izmer, fg = 'white', font = dFont )
lb_alarminterval.grid(row = row_num, column = 3, sticky=W, padx = 5)

ed_alarminterval = Entry(pn_izmer, width = dWd, font = dFont)
ed_alarminterval.grid(row = row_num, column = 4, padx= 5, pady= 5, sticky="nesw")



#== обработка кнопки "начать" 
def fn_start(ev):
    global flag_work
    global flag_calc
    global date_begin   #-- время начала счета  
    global interval     #-- длительность счета  в минутах
    
    if not flag_work:   #-- кнопка старта не была нажата
        flag_work = 1
        flag_calc = 1
        interval = 0
        date_begin = datetime.now() #-- заполнить поля начала счета
        stdate = date_begin.strftime('%H:%M:%S')

        ed_starttime.delete(0, END)
        ed_starttime.insert(END,stdate)

        ed_beginterval.delete(0, END)
        ed_beginterval.insert(END,stdate) 

        bt_start['text'] = u'завершить'     #-- изменить текст на кнопках
        bt_interval['text'] = u'остановить'
        
    else:               #-- нажата во время работы
        flag_work = 0   #-- остановить работу
        if flag_calc:
            date_end = datetime.now()
            delta_sec = (date_end - date_begin).total_seconds()        
            interval += delta_sec   #-- в секундах для демонстрации   
            
            flag_calc = 0   #-- остановить счет

        bt_start['text'] = u'начать'
        bt_interval['text'] = u'остановить'
    interval_min = interval / 60   
    st_interval = str(int(interval_min))    #-- вывести время работы
    ed_interval.delete(0, END)
    ed_interval.insert(END,st_interval)
        
bt_start.bind("<Button-1>", fn_start)   #-- подключить обработчик 
        
#== обработка кнопки "остановить" 
def fn_interval(ev):
    global flag_work
    global flag_calc
    global date_begin
    global interval
    
    if flag_work:
        if flag_calc:  # - нажали при счете - приостановить
            flag_calc = 0
            date_end = datetime.now()
            delta_sec = (date_end - date_begin).total_seconds()        
            interval += delta_sec

            interval_min = interval / 60   
            st_interval = str(int(interval))
            ed_interval.delete(0, END)
            ed_interval.insert(END,st_interval)

            date_begin = date_end
            bt_interval['text'] = u'продолжить'

        else:
            flag_calc = 1
            date_begin = datetime.now()
            stdate = date_begin.strftime('%H:%M:%S')

            ed_beginterval.delete(0, END)
            ed_beginterval.insert(END,stdate)

            bt_interval['text'] = u'остановить'
                   
bt_interval.bind("<Button-1>", fn_interval)     #-- подключить обработчик
        
#== обработка кнопки звонок "завести" 
def fn_alarm(ev):
    global flag_alarm       #-- признак заведенного звонка
    global date_alarm       #-- время срабатывания звонка
    
    if flag_alarm:
        flag_alarm = 0
        bt_alarm['text'] = u'завести'

    else:
        flag_alarm = 1

        try:                #-- защита на пустой интервал звонка 
            delta_alarm = 60*int(ed_alarminterval.get().strip())
        except:
            delta_alarm = 0

        date_alarm = datetime.now() + timedelta(seconds = delta_alarm)
        stdate = date_alarm.strftime('%H:%M:%S')
        ed_alarm.delete(0, END)
        ed_alarm.insert(END,stdate)

        bt_alarm['text'] = u'отключить'

bt_alarm.bind("<Button-1>", fn_alarm)
        
#== главная функция  
def fn_main():
    global flag_alarm
    global date_alarm

    date_cur = datetime.now()
    stdate_cur = date_cur.strftime('%H:%M:%S')
    ed_curtime.delete(0, END)
    ed_curtime.insert(END,stdate_cur)

    if flag_alarm:
        if date_cur > date_alarm:
            winsound.PlaySound('ringin.wav', winsound.SND_FILENAME)            
    root.after(5000, fn_main)
#== запуск в работу 
fn_main()
root.mainloop()

````

````python
def get_ip(host):
    args = ["ping","-n","1",host]
    CREATE_NO_WINDOW = 0x08000000
    proc=sp.Popen(args, stdout=sp.PIPE,
                creationflags = CREATE_NO_WINDOW)
    data=proc.communicate()
    txt=data[0].decode('cp866')
    ip_beg=txt.find('[')
    ip_end=txt.find(']')
    ip = txt[ip_beg+1:ip_end]
    return ip
print(get_ip('mail.ru'))    
    
````

#### py054 Питон применение Radiobutton и Checkbutton

````python
#coding: utf8
from tkinter import *
root = Tk()

lsp=[]
#-- левая панель с радиокнопками
fr_left = Frame(root)
fr_left.pack(side = 'left', fill = 'y')

##-- блок радиокнопок
rb_val = IntVar()
rb_val.set(3)

###-- записать в поле Entry значение радиокнопок
def calc_rb():
    global res
    res = rb_val.get()
    ed_val.delete(0,END)
    ed_val.insert(END, str(res))
    calc_bin(res)
        
###-- сформировать и вывести группу радиокнопок
for num in range(16):
    text = hex(num)
    rb = Radiobutton(fr_left, text=text, variable=rb_val, value = num, command = calc_rb)
    rb.pack(anchor=W)

#-- правая панель с чекбоксами 
fr_right = Frame(root)
fr_right.pack( side = 'right', fill = 'both', expand = 1)

##-- группа чекбоксов с надписями
###-- переменные для чекбоксов
var8 = IntVar()
var4 = IntVar()
var2 = IntVar()
var1 = IntVar()

###-- вывести в поле Entry и на радиокнопки значения из чекбоксов
def calc_cb():
    res = 8*var8.get() + 4*var4.get() + 2*var2.get() + var1.get()
    ed_val.delete(0,END)
    ed_val.insert(END, str(res))
    rb_val.set(res)

###-- сформировать и вывести группу чекбоксов
row_num = 0
Label(fr_right, text = 'binary: ').grid( row = row_num, column = 0)
row_num += 1
cb8 = Checkbutton(fr_right, text="8", variable=var8, command = calc_cb)
cb8.grid(row = row_num, column = 1, sticky=W)
cb4 = Checkbutton(fr_right, text="4", variable=var4, command = calc_cb)
cb4.grid(row = row_num, column = 2, sticky=W)
cb2 = Checkbutton(fr_right, text="2", variable=var2, command = calc_cb)
cb2.grid(row = row_num, column = 3, sticky=W)
cb1 = Checkbutton(fr_right, text="1", variable=var1, command = calc_cb)
cb1.grid(row = row_num, column = 4, sticky=W)

##-- массив чекбоксов
###-- сформировать и вывести заголовки для чекбоксов
row_num += 1
for num in range(8):
    Label(fr_right, text = str(7-num)).grid(row = row_num, column = num+1, sticky=W)

###-- вывести в поле Entry и на радиокнопки значения из массива чекбоксов
def calc_cb2():
    st = '0b'
    for el in ar_cbval:
        st = st+str(el.get())
    res = int(st,2)
    ed_val.delete(0,END)
    ed_val.insert(END, str(res) )
    rb_val.set(res)

###-- вывести в массив чекбоксов целое число
def calc_bin2( num ):
    st_num = bin(num+256)
    for num in range(8):
        bit = st_num[-1-num]
        ar_cbval[7-num].set( int(bit) )
    
###-- сформировать и вывести массив чекбоксов
row_num += 1
ar_cb = []
ar_cbval = []
for num in range(8):
    ar_cbval.append(IntVar())
    ar_cb.append(Checkbutton(fr_right, variable=ar_cbval[num], command = calc_cb2))
    ar_cb[num].grid( row = row_num, column = num + 1, sticky=W)
                 
##-- безимянная метка
row_num += 1
Label(fr_right, text = 'decimal: ').grid( row = row_num, column = 0, sticky=W)

##-- поле ввода
row_num += 1
ed_val = Entry(fr_right, width = 8)
ed_val.grid(row = row_num, column = 1, columnspan = 3, sticky=W)

row_num += 1

###-- вывести в группу чекбоксов целое число
def calc_bin( num ):
    ls_cb = [(8, var8),(4, var4),(2, var2),(1,var1)]
    for el in ls_cb:
        val = el[0]
        var = el[1]
        if num >= val:
            var.set(1)
            num -= val
        else:
            var.set(0)   

###-- вывести в чекбоксы и радиоккнопки целое число из Entry
def calc_dec():
    res = int(ed_val.get())
    calc_bin2(res)
    res = res % 16
    calc_bin(res)
    rb_val.set(res)
    
    
##-- кнопка
bt_dec = Button(fr_right, text = 'dec to binary', command = calc_dec)
bt_dec.grid(row = row_num, column = 1, columnspan = 3, sticky=W)

root.mainloop()
````

````python
#coding: utf8
import tkinter as tk 
import codecs

root = tk.Tk()
root.title("Кафе Солнышко")

lsp=[]

bt_wd = 15      # -- ширина кнопки
bg_cl = '#ff9988'    # - цвет формы
root['bg'] = bg_cl
#-- заголовок левой стороны
tk.Label(root, text = "Меню:", bg = bg_cl).grid(row = 0, column = 0)

#-- кнопка загрузки меню из файла
def loadMenu():
        global lsp
        fh = codecs.open('menu.txt','r','utf-8')
        ls_text = fh.readlines()
        fh.close()
        lsp = ls_text.copy()
        ls_prod = [st.rstrip() for st in ls_text]
        lx_Menu.delete(0, tk.END)
        for prod in ls_prod:
                lx_Menu.insert(tk.END, prod)

tk.Button(root, text = u"Загрузить", width = bt_wd, command = loadMenu).grid(row=0, column = 2)

#-- меню с полосой прокрутки
lx_Menu = tk.Listbox(root, width=50, height=10)
lx_Menu.grid(row=1, column=0, rowspan = 6, sticky=tk.N)

yscroll = tk.Scrollbar(command=lx_Menu.yview, orient=tk.VERTICAL)
yscroll.grid(row=1, column=1, rowspan = 6, sticky=tk.N+tk.S)
lx_Menu.configure(yscrollcommand=yscroll.set)

tk.Label(root, text = " ", bg = bg_cl).grid(row = 8, column = 0)

#-- заголовки правой стороны

tk.Label(root, text = "Заказ:", bg = bg_cl).grid(row = 0, column = 3)
tk.Label(root, text = "колич.", bg = bg_cl).grid(row = 0, column = 4)

#-- поля продкута и количества
ed_Prod = tk.Entry(root, width=50 )
ed_Prod.grid(row = 1, column = 3)
ed_Kolvo = tk.Entry(root, width=5 )
ed_Kolvo.grid(row = 1, column = 4)

#-- список заказа
lx_Zakaz = tk.Listbox(root, width=50, height=8)
lx_Zakaz.grid(row=2, column=3, rowspan = 6, sticky=tk.N)

#-- кнопка добавить в заказ
def addProd(ev = 0):
        st = ed_Prod.get() + " * " + ed_Kolvo.get()
        lx_Zakaz.insert( tk.END, st )

tk.Button(root, text = u"Добавить", width = bt_wd, command = addProd).grid(row=1, column = 5)

#-- кнопка изменить заказ
def editProd():
        lx_num = lx_Zakaz.curselection()[0]
        st = lx_Zakaz.get( lx_num )
        lx_Zakaz.delete( lx_num )

        pos = st.find('*')

        ed_Prod.delete(0, tk.END)
        ed_Prod.insert(0, st[0:pos].strip())

        ed_Kolvo.delete(0, tk.END)
        ed_Kolvo.insert(0, st[pos+1:].strip())

tk.Button(root, text = u"Изменить", width = bt_wd, command = editProd ).grid(row=2, column = 5)

#-- кнопка удалить позицию заказа
def delProd():
        try:
                lx_num = lx_Zakaz.curselection()[0]
                lx_Zakaz.delete( lx_num )
        except:
                pass
        
tk.Button(root, text = u"Удалить", width = bt_wd, command = delProd ).grid(row=3, column = 5)

#-- кнопка  очистить весь заказ
def clearZakaz():
        lx_Zakaz.delete(0, tk.END)

tk.Button(root, text = u"Очистить", width = bt_wd, command = clearZakaz ).grid(row=0, column = 5)

#-- кнопка  отсортировать заказ
def sortZakaz():
    ls_temp = list(lx_Zakaz.get(0, tk.END))
    ls_temp.sort()

    lx_Zakaz.delete(0, tk.END)

    for item in ls_temp:
        lx_Zakaz.insert(tk.END, item)

tk.Button(root, text = u"Сортировать", width = bt_wd, command = sortZakaz ).grid(row=5, column = 5)


#-- кнопка удалить позицию заказа
def saveZakaz():
    ls_temp = list(lx_Zakaz.get(0, tk.END))

    fh = open("zakaz.txt", "w")
    for st in ls_temp:
            fh.write( st + '\n')
    fh.close()

tk.Button(root, text = u"Сохранить", width = bt_wd, command = saveZakaz ).grid(row=6, column = 5)

#-- по Enter в Kolvo добавить строку в заказ
ed_Kolvo.bind('<Return>', addProd )

def selProd(ev):
    lx_num = lx_Menu.curselection()[0]
    st = lx_Menu.get( lx_num )

    ed_Prod.delete(0, tk.END)
    ed_Prod.insert(0, st)

    ed_Kolvo.delete(0, tk.END)
    ed_Kolvo.insert(0, "1")

#-- по Щеску мыши в меню заполнить поле продукта
lx_Menu.bind('<Double-1>', selProd)

root.mainloop()

def cod(st, dec_enc):
    # == перекодировать строку
    # -- dec_enc задает режим перекодировки
    # -- dec = "u","w","c","p" - декодировать
    # -- enc = "u","w","c","p" - кодировать
    # -- пример: st = cod(st,"up")
    dec = dec_enc[0]; enc = dec_enc[1]
    
    if dec == "u":
        st = st.decode("utf-8")
    if dec == "w":
        st = st.decode("cp1251")
    if dec == "c":
        st = st.decode("cp866")

    if enc == "u":
        st = st.encode("utf-8")
    if enc == "w":
        st = st.encode("cp1251")
    if enc == "c":
        st = st.encode("cp866")
    return st


st = lsp[0]
s2=st

# s3=cod(st,'uw')
print(lsp[0], s2.encode().decode())
````

#### py057 простой чат на UDP

````python
# simple UDP chat,  A.B.Glazov
#coding: utf8
#== подключить библиотеки
from tkinter import *
from socket import *
import threading
from time import *

#== сокет для приема сообщений
HOST_IN = ''
PORT_IN = 3000
BUFSIZE = 1024
SOCKADDR_IN = (HOST_IN,PORT_IN)
uServSock = socket(AF_INET,SOCK_DGRAM)
uServSock.bind(SOCKADDR_IN)

#== очередь для приема сообщений и флаг ее занятости
ls_in =[]
busy_in = 0

#== функция приема сообщений
def work_in():
        global ls_in
        global busy_in
        while True:
                data,addr = uServSock.recvfrom(BUFSIZE)
                loc_data = data.decode('cp1251') 
                st = 'receivedd from: ' +  addr[0] + ', data: ' +  loc_data

                while busy_in:
                        sleep(0.001)
                busy_in = 2
                ls_in.append(st)
                busy_in = 0

                sleep(0.001)

        uServSock.close()

#== поток приема сообщений
tr_in = threading.Thread( target = work_in)
tr_in.daemon = True
tr_in.start()

#== сокет для отправки сообщений
HOST_OUT = '127.0.0.1'
PORT_OUT = 3000
BUFSIZE = 1024
SOCKADDR_OUT = (HOST_OUT,PORT_OUT)
uCliSock = socket(AF_INET,SOCK_DGRAM)

#== очередь отправки сообщений и признак ее занятости
ls_out =[] 
busy_out = 0

#== сокет получателя сообщений
HOST_SRV = '10.1.0.190'
HOST_SRV = '192.168.0.112'

PORT_SRV = 3000
SOCKADDR_SRV = (HOST_SRV,PORT_SRV)

#== функция отправки сообщений
def work_out():
        global ls_out
        global busy_out
        global SOCKADDR_SRV
        while True:
                if not busy_out:
                        busy_out = 1
                        if len(ls_out) > 0:
                                st = ls_out.pop(0)
                                st = st.encode('cp1251') 
                                uCliSock.sendto(st, SOCKADDR_SRV)
                        busy_out = 0
                sleep(0.001)
        uCliSock.close()

#== поток отправки сообщений
tr_out = threading.Thread( target = work_out)
tr_out.daemon = True
tr_out.start()
        
#== основная программа: графическая часть 
root = Tk()

##== панель для ввода сообщений (нижняя)
pn_out = Frame(root, height = 50, bg = 'darkgreen')
pn_out.pack(side = 'bottom', fill = 'x')

###== метка для Ip адреса
Label(pn_out, text = '', bg='darkgreen' ).grid( row=0, column=0)
Label(pn_out, text = 'Ip_dest:' ).grid( row=1, column=0)

###== функция и поле ввода Ip адреса получателя
def set_ipout(event):
        global SOCKADDR_SRV
        ip_out = ed_ipout.get().strip()
        SOCKADDR_SRV = (ip_out,PORT_SRV)
        
ed_ipout = Entry(pn_out, width = 30)
ed_ipout.grid( row=1, column=1)
ed_ipout.bind('<Return>',set_ipout)

###== метка для поля отправляемого сообщения
Label(pn_out, text = 'message:' ).grid( row=1, column=3)

###== функция и поле ввода отправляемого сообщения
def send_mess(event):
    mess_out = ed_out.get().strip()
    if len(mess_out) > 0:
            tb_mess.insert(END, "\n>>> " + mess_out)
            ed_out.delete(0,END)
            ls_out.append(mess_out)
            
ed_out = Entry(pn_out, width = 30)
ed_out.grid( row=1, column=4)
ed_out.bind('<Return>',send_mess)

##== панель для вывода получекнных сообщений (основная)
pn_mess = Frame(root, height = 570, width = 700)
pn_mess.pack(side = 'top', expand = 1, fill = 'both')

###== текстовое поле для вывода получекнных сообщений
tb_mess = Text(pn_mess, height = 30, wrap=WORD )
tb_mess.pack(side = 'left', fill = 'both', expand = 1)

###== полоса прокрутки для текстового поля получекнных сообщений
sb_mess = Scrollbar(pn_mess)
sb_mess['command'] = tb_mess.yview
tb_mess['yscrollcommand'] = sb_mess.set
sb_mess.pack(side = 'right', fill = 'y')

#== главная функция, запускаемая в цикле 
def main():
    if len( ls_in ) > 0:
        mess_in  = ls_in.pop(0)
        tb_mess.insert(END, "\n<<< " + mess_in)
    root.after(30, main)
    
main()

#== запуск приложения в работу
root.mainloop()



````

#### py059 Крестики - нолики на Питоне


````python
# coding: utf-8
#-- игра крестики-нолики.  А.Б.Глазов
from tkinter import *
root = Tk()

#-- параметры поля
cell_size = 24
x_count = 30
y_count = 24

#-- признак окончания игры
game_over = 0

#-- массив результатов ходов
cells = []
for y_num in range(y_count):
    row = []
    for x_num in range(x_count):
        row.append(-1)
    cells.append(row)
        
#-- формирование канвы
canv_width = x_count * cell_size 
canv_height = y_count * cell_size 

canv = Canvas(root, width = canv_width, height = canv_height)
canv.pack()


#-- клеточки на канве
for x_num in range(x_count):
    for y_num in range(y_count):
        canv.create_line(0, y_num*cell_size, canv_width, y_num*cell_size)
        canv.create_line( x_num*cell_size, 0, x_num*cell_size, canv_height) 


#-- функция подсчета значений на линии
def calc_line(y_num, x_num, y_step, x_step):
    val = cells[y_num][x_num]
    
    val_count = 0
    for num in range(5):
        if cells[y_num + num * y_step][ x_num + num * x_step] == val:
            val_count += 1
        else:
            break
    for num in range(1,5):
        if cells[y_num - num * y_step][ x_num - num * x_step] == val:
            val_count += 1
        else:
            break
    return val_count
    
#-- функция подсчета значений на всех линиях около клетки
def calc_result(y_num, x_num):
    global game_over
    val_counts = []
    val_count = calc_line(y_num, x_num, 0, 1)
    val_counts.append(val_count) 
    val_count = calc_line(y_num, x_num, 1, 1)
    val_counts.append(val_count) 
    val_count = calc_line(y_num, x_num, 1, 0)
    val_counts.append(val_count) 
    val_count = calc_line(y_num, x_num, -1, 1)
    val_counts.append(val_count)
    val_count = max(val_counts)
    # print val_count
    if val_count >= 5:
        game_over = 1
        st = "END OF GAME "
        if cells[y_num][x_num] == 0:
            st = st + " 0 - WIN"
        else:
            st = st + " X - WIN"
        canv.create_text(280, 30, text = st, font = ("Arial", 30), tag = 'game')
        st = "press <Enter> for new game"
        canv.create_text(280, 80, text = st, font = ("Arial", 20), tag = 'game')
        

#-- функция отрисовки нолика
def draw_oval(event):
    global cells
    x_num = int (event.x / cell_size)
    y_num = int (event.y / cell_size)
    if cells[y_num][x_num] == -1 and not game_over:
        cells[y_num][x_num] = 0
        x_left = x_num * cell_size + 3
        y_top = y_num * cell_size + 3
        canv.create_oval(x_left, y_top, x_left + cell_size - 6, y_top + cell_size - 6, outline="#050", width = 3, tag = 'game') 
        calc_result(y_num, x_num)


#-- функция отрисовки крестика
def draw_krest(event):
    global cells
    x_num = int (event.x / cell_size)
    y_num = int (event.y / cell_size)

    if cells[y_num][x_num] == -1 and not game_over:
        cells[y_num][x_num] = 1
        x_left = x_num * cell_size + 3
        y_top = y_num * cell_size + 3
        canv.create_line(x_left, y_top, x_left + cell_size - 6, y_top + cell_size - 6, fill = 'blue', width = 3, tag = 'game')
        canv.create_line(x_left + cell_size - 6, y_top,  x_left, y_top + cell_size - 6, fill = 'blue', width = 3, tag = 'game') 
        calc_result(y_num, x_num)
        

#-- привязка к событиям мыши
canv.bind('<Button-1>', draw_oval)
canv.bind('<Button-3>', draw_krest)

#-- включить обработчик клавиатуры
def key_handl(event):
    # обработать нажатия клавиш
    global game_over
    global cells
    if event.keysym == "Return" and game_over:  # начать игру заново
        game_over = 0
        
        for y_num in range(y_count):
            for x_num in range(x_count):
                cells[y_num][x_num] = -1

        canv.delete("game")        


canv.bind("<KeyPress>", key_handl)
canv.focus_set()

#-- запустить игру
root.mainloop()
````
