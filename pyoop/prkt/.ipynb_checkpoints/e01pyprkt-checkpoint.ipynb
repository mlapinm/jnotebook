```python
import os
import lib01

from bs4 import BeautifulSoup

def getListDirHtml():
    listDir = os.listdir()
    listDirHtml = []
    for i in listDir:
        if i.find('html')>=0:
            listDirHtml.append(i)
    return listDirHtml

listDirHtml = getListDirHtml()
strHtml=''
strHtml += '''	<table>
		'''
for i in listDirHtml:
    strHtml +='''<tr>
			<td><a href="'''
    strHtml += i
    strHtml += '''">'''
    strHtml += i
    strHtml += '''</a></td>
			<td></td>
			<td></td>
			<td></td>
			<td></td>
			<td></td>
		</tr>
		'''    
strHtml += '''	</table>'''

lib01.writeHTML('index2.html', strHtml)
print('ok')

soup = BeautifulSoup(strHtml,'lxml')

print(soup.head)
print(listDirHtml)
print(strHtml)
# print(listDir)
```

```python
#!/usr/bin/python3
 
from bs4 import BeautifulSoup
 
with open("e01emtab.html", "r") as f:
    
    contents = f.read()
 
    soup = BeautifulSoup(contents, 'lxml')
 
    root = soup.table.tr
    
    root_childs = [e.name for e in root.children if e.name is not None]
    print(root_childs)
```

#### BeautifulSoup метод find(), поиск элементов по id

```python
from bs4 import BeautifulSoup
 
with open("e01emtab.html", "r") as f:
    
    contents = f.read()
    newTag = soup.new_tag('li')
    newTag.text = "111"
    soup = BeautifulSoup(contents, 'lxml')
    tag = soup.find("li")
    tag.replace_with(newTag)
    print(contents)
    
    for tag in soup.find_all("li"):
        print("{0}: {1}".format(tag.name, tag.text))
    
```

#### BeautifulSoup метод append() добавление нового HTML-тега

```python
from bs4 import BeautifulSoup
 
with open("e01emtab.html", "r") as f:
    
    contents = f.read()
 
    soup = BeautifulSoup(contents, 'lxml')
 
    newtag = soup.new_tag('li')
    newtag.string='66'
 
    ultag = soup.ul
    
    ultag.append(newtag)
    
    print(ultag.prettify())
```

```python
%%writefile lib01.py
#lib01.py
import codecs

string0 = '''\
<!DOCTYPE html>
<html lang="en">
<head>
	<meta charset="UTF-8">
	<title>Document</title>
</head>
<body>
'''
string2 = '''\
</body>
</html>
'''

def writeHTML(name, string1):
    global string0, string2
    with codecs.open(name, 'w', 'utf-8') as file:
        file.write(string0 + string1 + string2)


def writeFile(name, string1):
    with codecs.open(name, 'w', 'utf-8') as file:
        file.write(string1)

def readFile(name):
    string1 = ''
    with codecs.open(name, 'r', 'utf-8') as file:
        string1 = file.read()
    return string1
```
