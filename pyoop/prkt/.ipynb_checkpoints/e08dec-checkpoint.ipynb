```python
import codecs
import os
import re

CODEC_FROM = "cp1251"
CODEC_TO = "utf-8"


class ClDecode:
    MASK_FILES = ['*.cpp','*.h','*.txt',]
    DIR_TO = '11'
    
    def __init__(self, builder):
        self.MASK_FILES=builder.MASK_FILES
        self.DIR_TO=builder.DIR_TO
        
    def mListFiles(self,):
        return os.listdir('.')

    def getMask(self,s):
        '''
        s = '*.cpp'
        return: 'cpp'
        '''
        pattern = r'[*].(\w+)$' 
        string = s
        match = re.search(pattern, string)
        s2 = None
        if match:
            if len(match.groups())>0:
                s2 = match.group(1)
        return s2

    def mFindMask(self,s,p):
        '''
        s='file.cpp'
        p ='cpp'
        '''
        pattern = r'.+\.'+ p + '$'
        string = s
        match = re.search(pattern, string)
        s2 = None
        if match:
            s2 = match[0]
        return s2

    def mFrToDecode(self,file):
        '''
        f1 = 'e08dec.txt'
        '''
        if os.path.isdir(self.DIR_TO) == False:
            os.mkdir(self.DIR_TO)
        fileTo = self.DIR_TO + file
        sform1 = "utf-8" #ANSI
        st = ""
        with codecs.open(file,"r",CODEC_FROM) as fh :
            st = fh.read()

        with codecs.open(fileTo,"w",CODEC_TO) as fh :
            fh.write(st)

    def mMethod(self,):
        files = self.mListFiles()
        files2 = []
        for i in files:
                for j in self.MASK_FILES:
                    m = self.getMask(j)
                    s = self.mFindMask(i,m)
                    if s:
                        files2.append(s)
        for i in files2:
            print(i)
            self.mFrToDecode(i)
        print(files2)
   
class Builder:
    MASK_FILES = ['*.cpp','*.h','*.txt',]
    DIR_TO = '..\\tmp1\\'
    
    def setMaskFiles(self, maskFiles):
        self.MASK_FILES = maskFiles
        return self
    
    def setDirTo(self, dirTo):
        self.DIR_TO = dirTo
        return self
    
    def build(self):
        return ClDecode(self)
    
builder = Builder()
builder.setMaskFiles(['*.cpp','*.h','*.txt',])
builder.setDirTo('..\\tmp1\\')
clDecode=builder.build()
clDecode.mMethod()

print(clDecode.MASK_FILES)
print(clDecode.DIR_TO)

```
