```python
%%writefile prep.py
# coding: utf-8
# вспомогательные функции для формирования тестов
import re
import codecs

# записать текст в файл
def write_text( fil_name, text ):
    fh = codecs.open( fil_name, 'w','utf-8' )
    fh.write(text)
    fh.close( )

# прочитать текст из файла    
def read_text(fil_name ):
    fh = codecs.open( fil_name ,'r','cp1251')
    text = fh.read( )
    fh.close( )
    return text

# удалить непарный тэг     
def cut_tag( tag, text ):
    st = "<" + tag + "[^>]*?>"
    rex = re.compile( st )
    return re.sub( rex, "", text)

# очистить параметры тэгв
def clear_tagparams( tag, text ):
    st = "<" + tag + "[^>]*?>"
    rex = re.compile( st )
    return re.sub( rex, "<" + tag + ">" , text)

# удалить тэг и его завершение
def cut_tagpair( tag, text ):
    st = "<" + tag + "[^>]*?>"
    rex = re.compile( st )
    text = re.sub( rex, "", text)
    rex = re.compile( r"<\/"+tag+">" )
    return re.sub( rex, "", text)

# удалить полностью тэг с содержимым до завершающего включительно
def cut_fulltag( tag, text ):
    st = "<" + tag + r"[\s\S]*?<\/" + tag + ">"
    rex = re.compile( st )
    return re.sub( rex, "", text)

# извлечь содержимое тега
def get_tagcontent(tag, text):
    st = "<"+tag+"[^>]*?>([\s\S]*?)<\/"+tag+">"
    rex = re.compile(st)
    res = re.search( rex, text)
    return res.group(1)
            
# очистить группы неразрывных пробелов
def clear_nbsp( text ):
    rex = re.compile( "(&nbsp;)+" )
    return re.sub( rex, " ", text)

# очистить группы пустых строк
def clear_eol( text ):
    rex = re.compile( "[\n\r]+" )
    return re.sub( rex, "\n", text)
```

```python
#coding: cp1251
import re
import prep

#--функци¤ обработки одного блока вопросов    
def make_form( quest_num, ls_questdelim, ls_quest ):
    quest = ls_quest[ quest_num ]
    quest_delim = ls_questdelim[quest_num -1]
    
    rex_answ = re.compile("<p>\S\)")
    ls_answdelim = re.findall(rex_answ, quest)
    ls_answ = re.split(rex_answ, quest)
    
    quest_head = quest_delim + ls_answ[0].strip()
    form = '\n<FORM NAME ="f'+ str(quest_num) + '" class = "answers"> ' + quest_head
    for answ_num in range(1, len(ls_answ)):
        answ_delim = ls_answdelim[ answ_num-1 ]
        answ = ls_answ[answ_num ].strip()
        form += '\n    '+ answ_delim + '<INPUT TYPE="radio" NAME="rd1" value = '
        form += str(answ_num) +'>'+answ
    form += '\n</FORM>'
    form +='\n<hr color = blue align = center width = 50%>\n'
    return form
    
#-- прочитать исходный файл
# fil_name = input('input file name:')
fil_name = 'alg1.htm'
text = prep.read_text(fil_name)

#-- удалить весь тэг style с содержимым в заголовке
head = prep.get_tagcontent('head', text)
head = prep.cut_fulltag('style', head)

#-- удалить все неразрывные пробелы из body
text = prep.get_tagcontent('body', text)
text = prep.clear_nbsp( text)

#-- удалить все парные тэги span div
text = prep.cut_tagpair('span', text)
text = prep.cut_tagpair('div', text)

#-- удалить непарные тэги
text = prep.cut_tag('br', text)

#-- стереть параметры в тэге р
text = prep.clear_tagparams('p', text)

#-- стереть лишние концы строк
text = prep.clear_eol( text)

#-- сформировать список заданий
rex_quest = re.compile("<p>\d+\.")
ls_questdelim = re.findall(rex_quest, text)
ls_quest = re.split(rex_quest, text)

#-- сформировать новое body
text = ls_quest[ 0 ]
for form_num in range(1, len(ls_quest)):
    form =  make_form( form_num, ls_questdelim, ls_quest )
    text += form

#--дописать подвал страницы
footer ='''
\n<script src="test_calc.js"></script>

<p></p>
   <FORM NAME ="formrez">
      <P  ALIGN = CENTER >
      <INPUT TYPE="text" NAME="test_res" id="test_res" SIZE=80 VALUE=" " style="font-size:20px">
      <br>     
      <INPUT TYPE="button" VALUE="–езультат" onClick="CalcMark();" style="font-size:20px">
   </FORM>\n
'''
text += footer

#--сформировать полный результат
new_page = "<html><head>" + head +"</head>"
new_page +="<body> "+ text +"</body></html>"

#--записать результат в файл htm 
prep.write_text( "res"+fil_name, new_page)
```
