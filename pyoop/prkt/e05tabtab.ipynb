```python
import codecs
import clipboard

st = ""
with codecs.open("e05detr.txt","r","utf-8") as fh :
    st = fh.read();
st = st.replace("<","&lt;")
st = st.replace(">","&gt;")
st = st.replace("\"","&quot;")

lst = st.split("\n")
lst = [s for s in lst if len(s.strip())>0]
lst2 = []
lks = []
stt=[]
k=0
lks.append("<html>")
lks.append("<table>")
for st in lst:
    lst2 =st.split("\t")
    lenst = len(lst2)
    s1="    <td>"
    s2="    <td>"
    if lenst == 1:
        s1+= "_"
        s2+= lst2[0].strip()
    if lenst == 2:
        s1+= lst2[0].strip()
        s2+= lst2[1].strip()
    s1 += "</td>"
    s2 += "</td>"
    lks.append("<tr>")
    lks.append(s1)
    lks.append(s2)
    lks.append("</tr>")
lks.append("</table>")
lks.append("</html>")
st = "\n".join(lks)
print(st)    
clipboard.copy(st)  
```
