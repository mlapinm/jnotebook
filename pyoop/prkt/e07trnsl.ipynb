```python
import codecs

FILE_SRT = "e07en01r.txt"
FILE_EN = "e07en01en.txt"
FILE_NUM_EN = "e07en01numen.txt"

def mFrSrt(file_write,num_is):
    st = ""
    ls = []
    bom = True
    with open(FILE_SRT,"rb") as fh:
        lb = [b'\xef',b'\xbb',b'\xbf']
        for i in range(3):
            if lb[i] != fh.read(1):
                bom = False
                break
    with codecs.open(FILE_SRT,"r","utf-8") as fh:
        if bom:
           fh.read(1)
        ls = [i.strip() for i in fh if len(i.strip())>0 ]
    ls = [i for i in ls if i.find("-->")==-1]
    # ls = [i for i in ls if not i.isdigit()]
    ls1 = ls.copy()
    ls = []
    symb = ['.','!','?']
    num = 0
    for i in ls1:
        if i.isdigit():
            if len(st)==0:
                num = int(i)
        else:
            bsymb = False
            for s in symb:
                if(i[-1]==s):
                    bsymb=True
            probel = ' '
            if len(st)==0:
                probel = ''
            if not bsymb:        
                st += probel + i
            else:
                if num_is:
                    ls.append(str(num))
                ls.append(st + probel + i)
                st = ""

    st = "\n".join(ls)
    print(len(ls1),len(ls),ls)
    with codecs.open(file_write,"w","utf-8") as fh:
        fh.write(st)
mFrSrt(FILE_EN,num_is=False)
mFrSrt(FILE_NUM_EN,num_is=True)
```

```python
import requests
import json

def translate(world):
    payload = {
       'key': r'trnsl.1.1.20190909T123034Z.91e60547a2d5b3f5.bbf98325fde572ca3205b83b31a9034d9ff08f94',
       'text':'additional',
       'lang':'en-ru'
    }
    payload['text']=world
    url = 'https://translate.yandex.net/api/v1.5/tr.json/translate'
    r = requests.post(url, data=payload)
    answer = json.loads(r.text)
    word = answer['text'][0] if len(answer['text'])>0 else ''
    return word

st= ""
st=translate("Oh, the... there's a 7:15 at the ArcLight.")
st
```

```python
#FILE_EN
FILE_RU = "e07en01ru.txt"

def mTranslate():
    ls = []
    with codecs.open(FILE_EN,"r","utf-8") as fh:
        ls = [i.strip() for i in fh if len(i.strip())>0 ]
    lsr =[]
    for s in ls:
        lsr.append(translate(s))
        
    st = "\n".join(lsr)    
    with codecs.open(FILE_RU,"w","utf-8") as fh:
        fh.write(st)
    print(lsr)

mTranslate()
```
