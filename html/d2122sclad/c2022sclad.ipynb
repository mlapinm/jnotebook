```python
import re
import json

class Lines:
    '''
    nextName, readLines, read, write, writeLines, mIntervals, , ,
    '''
    name = ''
    nameMask = ''
    lines = []
    text = ""
    lines2 = [] # копия входных данных
    objSpls = []
    objElms = []
    enRuObjLines = []
    objText = ''
    enLines = []  
    ruLines = []
    enRuLines = []

    def __init__(self, name):
        self.name = name
        self.nameMask = self.nextName(self.name)



    def nextName(self, name):
        name2 = ''
        patten = r'^\w\d\d\d\d'
        match = re.search(patten, name)
        name2 = match[0] if match else 'w000'
        return name2

    def readLines(self, name, items):
        with open(self.name, 'r', encoding='utf8') as file:
            for line in file:
                line2 = line.strip()
                len(line2) == 0 or items.append(line2)
       

    def read(self, name):
        self.lines = []
        self.readLines(name, self.lines)
        self.text = '\n'.join(self.lines)
        return self.text

    def write(self, name, text):
        with open(name, 'w', encoding='utf8') as file:
            file.write(text)

    def writeLines(self, name, items):
        text = '\n'.join(items)
        with open(name, 'w', encoding='utf8') as file:
            file.write(text)


    def mRu(self, text):
        ru = False
        ru = True if re.search('^[а-яА-Я]', text) else False
        return ru
        pass

    def mEn(self, text):
        en = False
        en = True if re.search('^[a-zA-Z]', text) else False
        return en
        pass



 
    def mObjText(self, objl):
        k = 0
        text = json.dumps([ob for ob in objl], indent=2, ensure_ascii=False)
        return text

    def mTransform(self, l1, l2):

        l0 = []

        for e in l1:
            patten = r'^\d\d[a-zA-Z]'
            match = re.search(patten, e)
            if match:
                l0 = []
                l0.append(match[0])
            else:
                patten = r'^(\d\d\.\d\d)'
                t = re.sub(patten, '', e)
                l01 = l0.copy()
                l01.append(t.strip())
                l2.append(l01)

        pass

    def mObjSpls(self, l1, l2):
        obj0 = {"quanity" : "", "name" : "", "box" : ""}
        for e in l1:
            obj = obj0.copy()
            obj["name"] = e[1]
            obj["box"] = e[0]
            obj["quanity"] = 1
            l2.append(obj)
        pass

    def mObjElms(self, l1, l2):
        obj0 = {"name" : "", 
        "tags" : ["", "", ""], 
        "description" : ["", "", ""], 
        "image" : ["", "", ""], 
        "links" : ["", "", ""]}
        for e in l1:
            obj = obj0.copy()
            obj["name"] = e[1]
            obj["tags"] = ["aaa", "", ""]
            l2.append(obj)
        pass



lines = Lines('c2022a.txt')
lines.read(lines.name)
lines.mTransform(lines.lines, lines.lines2)
lines.mObjSpls(lines.lines2, lines.objSpls)
lines.mObjElms(lines.lines2, lines.objElms)
# print(lines.objSpls)
objText = lines.mObjText(lines.objSpls)
lines.write(lines.nameMask + 'b1.txt', objText)
objText = lines.mObjText(lines.objElms)
lines.write(lines.nameMask + 'b2.txt', objText)

# lines.writeLines(lines.nameMask + 'c1.txt', lines.lines)

```
